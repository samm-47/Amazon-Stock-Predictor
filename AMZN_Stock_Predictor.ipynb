```python
import yfinance as yf
```

```python
AMZN = yf.Ticker("^GSPC")
```

```python
AMZN = AMZN.history(period="max")
```

```python
AMZN
```

```python
AMZN.index
```

```python
AMZN.plot.line(y="Close", use_index=True)
```

```python
del AMZN["Dividends"]
del AMZN["Stock Splits"]
```

```python
AMZN["Tomorrow"] = AMZN["Close"].shift(-1)
```

```python
AMZN
```

```python
AMZN["Target"]=(AMZN["Tomorrow"]>AMZN["Close"]).astype(int)
```

```python
AMZN
```

```python
AMZN = AMZN.loc["1990-01-01":].copy()
```

```python
AMZN
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=400, min_samples_split=100,random_state=1)

train = AMZN.iloc[:-100]
test = AMZN.iloc[-100:]

predictors = ["Close","Volume","Open","High","Low"]
model.fit(train[predictors], train["Target"])
```

```python
RandomForestClassifier(min_samples_split=100, random_state=1)
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds=pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds=model.predict(test[predictors])
    preds=pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions=[]

    for i in range(start, data.shape[0], step):
      train = data.iloc[0:i].copy()
      test = data.iloc[i:(i+step)].copy()
      predictions = predict(train, test, predictors, model)
      all_predictions.append(predictions)
      return pd.concat(all_predictions)
```

```python
predictions = backtest(AMZN, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
  rolling_averages = AMZN.rolling(horizon).mean()

  ratio_column = f"Close_Ratio_{horizon}"
  AMZN[ratio_column] = AMZN["Close"] / rolling_averages["Close"]

  trend_column = f"Trend_{horizon}"
  AMZN[trend_column] = AMZN.shift(1).rolling(horizon).sum()["Target"]

  new_predictors += [ratio_column, trend_column]
```

```python
AMZN = AMZN.dropna()
```

```python
AMZN
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50,random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds=model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds=pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(AMZN, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
